```python
from tkinter import *
from tkinter import messagebox
from PIL import Image, ImageTk



class Student:
    def __init__(self, root):
        self.root = root
        self.root.geometry('1280x853+0+0')

        # Load the background image
        try:
            bg_image = Image.open('bg.jpg')
            self.background_image = ImageTk.PhotoImage(bg_image)
            bg_label = Label(root, image=self.background_image)
            bg_label.place(x=0, y=0, relwidth=1, relheight=1)
        except FileNotFoundError:
            print("Error: Background image 'bg.jpg' not found.")

        # Load the logo image
        try:
            logo_image = Image.open('user.png')
            self.logo_image = ImageTk.PhotoImage(logo_image)
            logo_label = Label(root, image=self.logo_image)
            logo_label.place(relx=0.5, rely=0.3, anchor=CENTER)  # Center the logo vertically
        except FileNotFoundError:
            print("Error: Logo image 'user.png' not found.")

        # Load username and password icon images
        self.username_image = PhotoImage(file='username.png')
        self.password_image = PhotoImage(file='password.png')

        # Create username label with icon
        username_label = Label(root, image=self.username_image, text='Username', compound=LEFT, font=('times new roman', 20, 'bold'))
        username_label.place(relx=0.35, rely=0.45, anchor=W)  # Adjust placement relative to window size

        # Create username entry
        self.username_entry = Entry(root, font=('times new roman', 20, 'bold'))
        self.username_entry.place(relx=0.5, rely=0.45, anchor=W)  # Adjust placement relative to window size

        # Create password label with icon
        password_label = Label(root, image=self.password_image, text='Password', compound=LEFT, font=('times new roman', 20, 'bold'))
        password_label.place(relx=0.35, rely=0.55, anchor=W)  # Adjust placement relative to window size

        # Create password entry with show='*' for hiding characters
        self.password_entry = Entry(root, font=('times new roman', 20, 'bold'), show='*')
        self.password_entry.place(relx=0.5, rely=0.55, anchor=W)  # Adjust placement relative to window size

        # Create login button with command set to the login function
        login_button = Button(root, text='Login', font=('times new roman', 20, 'bold'), width=15, fg='white', bg='cornflowerblue',
                              activebackground='cornflowerblue', activeforeground='white', cursor='hand2', command=self.login)
        login_button.place(relx=0.5, rely=0.65, anchor=CENTER)  # Adjust placement relative to window size

    def login(self):
        if self.username_entry.get() == '' or self.password_entry.get() == '':
            messagebox.showerror('Error', 'Fields cannot be empty')
        elif self.username_entry.get() == 'Sushanth' and self.password_entry.get() == '1234':
            messagebox.showinfo('Success', 'Welcome')
            self.root.destroy()
            # Close the main window
            self.open_new_window()
            import Sms# Call method to open new window
        else:
            messagebox.showerror('Error', 'Please enter correct details')

    def open_new_window(self):
        # Create an instance of the new window and display it
        self.new_window = Toplevel()
        Sms.Student(self.new_window)  # Assuming Student is defined in sms module

if __name__ == '__main__':
    root = Tk()
    obj = Student(root)
    root.mainloop()


```
